```python
import easyocr

# Create reader
reader = easyocr.Reader(
    ['en'],   # English + Arabic
    gpu=False
)

# Read image
results = reader.readtext("src/numb.jfif")

# Print results
for bbox, text, confidence in results:
    print(f"Text: {text}")
    print(f"Confidence: {confidence:.3f}")
    print()
```
