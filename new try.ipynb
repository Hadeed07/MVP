```python
import easyocr
import os
import cv2

reader_en = easyocr.Reader(['en'])
reader_ur = easyocr.Reader(['ur'])
```

```python
def extract_text_with_rotation(image_path, reader_en):
    img = cv2.imread(image_path)
    best_text = ""
    best_confidence = 0

    for angle in [0, 90, 180, 270]:
        if angle == 0:
            rotated = img
        elif angle == 90:
            rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
        elif angle == 180:
            rotated = cv2.rotate(img, cv2.ROTATE_180)
        else:
            rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

        detections = reader_en.readtext(rotated)
        if detections:
            avg_conf = sum(d[2] for d in detections) / len(detections)
            text = " ".join([d[1] for d in detections])
            if avg_conf > best_confidence:
                best_confidence = avg_conf
                best_text = text

    return best_text, best_confidence


def process_folder(crop_folder, reader):
    results = {}
    for filename in os.listdir(crop_folder):
        if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
            path = os.path.join(crop_folder, filename)
            text, confidence = extract_text_with_rotation(path, reader)
            results[filename] = {"text": text, "confidence": confidence}
            print(f"{filename}: '{text}' (confidence: {confidence:.2f})")
    return results


# Run it on your book_crops folder
extracted = process_folder("outputs_2\crops", reader_en)
```
